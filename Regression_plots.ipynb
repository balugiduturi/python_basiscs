```python
import seaborn as sns
%matplotlib inline
```

```python
tipsData = sns.load_dataset('tips')
```

```python
tipsData.head()
```

```python
regress = sns.lmplot(x='total_bill', y='tip', data=tipsData)
```

### If we want to plot with categories

```python
regress = sns.lmplot(x="total_bill", y = 'tip', data=tipsData, hue = 'sex')
```

### Chage the marker size & style

```python
regress = sns.lmplot(x='total_bill', y='tip', data=tipsData, hue='sex', scatter_kws={'s':100}, markers=['v', 'x'])
```
